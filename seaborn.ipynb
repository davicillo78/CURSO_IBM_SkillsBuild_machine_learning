```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas.head(10)  # Muestra las primeras filas del DataFrame
```

```python
#histograma para visualizar una sola variable
sns.histplot(propinas["total_bill"], bins=30, kde=True)
```

```python
#visualizacion de dos variables
sns.jointplot(data=propinas, x="total_bill", y="tip")
#grafico de distribucion junto a histograma de ejex x, y
```

```python
sns.jointplot(data=propinas, x="total_bill", y="tip", kind="hex")
```

```python
sns.jointplot(data=propinas, x="total_bill", y="tip", kind="reg")
#regresion lineal
```

```python
sns.jointplot(data=propinas, x="total_bill", y="tip", kind="kde")
#kde para visualizar la densidad de probabilidad conjunta
```

### Comparación de todas las variables numéricas ###

```python
sns.pairplot(data=propinas)
#matriz de dispersión para visualizar todas las variables
```

### Comparando variables numéricas y no numéricas (categóricas) ###

```python
sns.pairplot(propinas, hue="sex")
#matriz de dispersión para visualizar todas las variables, diferenciando por sexo
```

```python
sns.pairplot(propinas, hue="size", palette="husl")
#matriz de dispersión para visualizar todas las variables, diferenciando por tamaño de grupo
```

```python
sns.rugplot(data=propinas, x="total_bill", height=0.1)
#visualiza la densidad de probabilidad de una sola variable
```

### Gráficos para columnas categóricas ###

```python
propinas = sns.load_dataset("tips")
propinas.head(10)  # Muestra las primeras filas del DataFrame
```

```python
sns.barplot(data=propinas, x="sex", y="total_bill", palette="husl")
#visualiza la media de total_bill por sexo
```

```python
sns.countplot(data=propinas, x='sex', palette="husl")
#visualiza la cantidad de observaciones por sexo
```

```python
sns.countplot(data=propinas, x='smoker', palette="husl")
#visualiza la cantidad de observaciones por fumador
```

### Comparación de variable categórica con variable no numérica ###

```python
sns.boxplot(data=propinas, x="day", y="total_bill", palette="husl")
#visualiza la mediana de total_bill por dia de la semana
```

```python
sns.boxplot(data=propinas, x="day", y="total_bill", hue= 'smoker', palette="husl")
#visualiza la mediana de total_bill por dia de la semana, diferenciando por fumador
```

```python
sns.violinplot(data=propinas, x="day", y="total_bill", hue= 'smoker', palette="husl")
#visualiza la mediana de total_bill por dia de la semana, diferenciando por fumador
```

```python
sns.violinplot(data=propinas, x="day", y="total_bill", hue= 'smoker', palette="husl", split=True)
#visualiza la mediana de total_bill por dia de la semana, diferenciando por fumador, en un solo grafico
```

```python
sns.stripplot(data=propinas, x="day", y="total_bill", palette="husl", hue= 'sex')
#visualiza la mediana de total_bill por dia de la semana, diferenciando por sexo
```

```python
#para evitar que se solapen los puntos del gráfico de distribucion
sns.swarmplot(data=propinas, x="day", y="total_bill", palette="husl", hue= 'sex')
#visualiza la mediana de total_bill por dia de la semana, diferenciando por sexo
```

```python
#dos diagramas superpuestos
sns.violinplot(data=propinas, x="day", y="total_bill", palette="husl")
sns.swarmplot(data=propinas, x="day", y="total_bill", palette="husl")
#visualiza la mediana de total_bill por dia de la semana, diferenciando por sexo
```
