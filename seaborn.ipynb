### Mapas de calor ###

```python
import seaborn as sns
```

```python
vuelos = sns.load_dataset("flights")
vuelos.head(10)  # Muestra las primeras filas del DataFrame
```

```python
vuelos_matrix = vuelos.pivot_table(index="month", columns="year", values="passengers")
#pivotar el DataFrame para que las filas sean los meses, las columnas los años y los valores el número de pasajeros
vuelos_matrix.head(10)  # Muestra las primeras filas del DataFrame
```

```python
sns.heatmap(vuelos_matrix, cmap="viridis") #cmap es el mapa de colores
#visualiza la matriz de pasajeros por mes y año
```

```python
#separacion de los cuadrados del mapa de calor
sns.heatmap(vuelos_matrix, cmap="viridis", linecolor="white", linewidths=1)
#visualiza la matriz de pasajeros por mes y año, con separacion de los cuadrados
```

### Gráficos de cuadrícula (grids) ###

```python
import matplotlib.pyplot as plt
```

```python
flores = sns.load_dataset("iris")
flores.head(10)  # Muestra las primeras filas del DataFrame
```

```python
sns.pairplot(flores, hue="species", palette="husl")
#matriz de dispersión para visualizar todas las variables, diferenciando por especie
```

```python
# Pairgrid
graficos = sns.PairGrid(flores, palette="husl")
graficos.map(plt.scatter)
#graficos pairgrid en modelo scatter
```

```python
#configurar los gráficos del pairplot con la opción pairgrid
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.histplot) #diagonal con graficos histograma
```

```python
graficos = sns.PairGrid(flores)
graficos.map_offdiag(sns.scatterplot) #fuera de la diagonal con graficos scatter
```

```python
graficos = sns.PairGrid(flores)
graficos.map_lower(sns.kdeplot) #diagonal inferior izquierda con graficos kde
```

```python
graficos = sns.PairGrid(flores)
graficos.map_upper(sns.lineplot) #diagonal superior derecha con graficos linea
```

### Gráficos de regresión ###

```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas.head(10)  # Muestra las primeras filas del DataFrame
```

```python
sns.lmplot(data=propinas, x="total_bill", y="tip")
#regresion lineal de total_bill y tip
```

```python
#misma gráfica separada por sexo
sns.lmplot(data=propinas, x="total_bill", y="tip", hue= 'sex')
#regresion lineal de total_bill y tip, diferenciando por sexo
```

```python
#configurando los markers y el tamaño de los puntos
sns.lmplot(data=propinas, x="total_bill", y="tip", hue='sex', markers=["o", "v"], scatter_kws={"s": 75})
#con markers configuramos los tipos de marcadores (funciona como una lista)
#con scatter_kws configuramos el tamaño de los marcadores (funciona como un diccionario)
```

```python
#mismo gráfico separando en dos gráficos por sexo separación en columnas
sns.lmplot(data=propinas, x="total_bill", y="tip", col= 'sex')
```

```python
#mismo gráfico separando en dos gráficos por sexo separación en filas
sns.lmplot(data=propinas, x="total_bill", y="tip", row= 'sex')
```

```python
#modificas las dimensiones del gráfico con aspect y size
sns.lmplot(data=propinas, x="total_bill", y="tip", aspect= 1.5, height= 20)
```

### Estilos y colores ###

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas.head(10)  # Muestra las primeras filas del DataFrame
```

```python
#cambio de estilo de fondo de la gráfica
sns.set_style("darkgrid")  # Cambia el estilo de la gráfica a "darkgrid"
sns.countplot(data=propinas, x='sex', hue = 'sex')
#visualiza la cantidad de observaciones por sexo
```

```python
#cambio de estilo del marco exteriror de la gráfica
sns.set_style("white")  # Cambia el estilo de la gráfica a "whitegrid"
sns.countplot(data=propinas, x='sex', hue = 'sex')
#visualiza la cantidad de observaciones por sexo
sns.despine(left=True, bottom=True)  # Elimina el marco exterior de la gráfica
```

```python
#cambiar el tamaño del gráfico
plt.figure(figsize=(10, 6))  # Cambia el tamaño de la figura a 10x6 pulgadas
sns.set_style("darkgrid")  # Cambia el estilo de la gráfica a "darkgrid"
sns.countplot(data=propinas, x='sex', hue = 'sex')
#visualiza la cantidad de observaciones por sexo
```

```python
#cambiar el tamaño de las letras
sns.set_context("notebook", font_scale=1.5)  # Cambia el tamaño de la fuente a 1.5 veces el tamaño predeterminado
sns.set_style("darkgrid")  # Cambia el estilo de la gráfica a "darkgrid"
sns.countplot(data=propinas, x='sex', hue = 'sex')
#visualiza la cantidad de observaciones por sexo
```

```python
#cambiar los colores de la gráfica
sns.set_style("darkgrid")  # Cambia el estilo de la gráfica a "darkgrid"
sns.countplot(data=propinas, x='sex', hue = 'sex', palette='Set2')
#visualiza la cantidad de observaciones por sexo
```

```python
#cambiar los colores de la gráfica de forma manual
colores = ["grey", "blue"]  # Define una lista de colores personalizados
sns.set_style("darkgrid")  # Cambia el estilo de la gráfica a "darkgrid"
sns.countplot(data=propinas, x='sex', hue = 'sex', palette=colores)
#visualiza la cantidad de observaciones por sexo
```
